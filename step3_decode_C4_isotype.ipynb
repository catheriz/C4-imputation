```python
# Read the dosages and INFO data from the file
dosages_info = pd.read_csv('processed_data.txt', sep='\t')

# Initialize a dictionary to store C4 isotype counts for each sample
c4_isotype_counts = {sample: {'C4A': 0, 'C4B': 0, 'C4L': 0, 'C4S': 0} for sample in dosages_info.columns[3:]}

# Calculate the number of C4 isotypes for each sample
for index, row in dosages_info.iterrows():
    haplotype = row['ALT']
    dr2 = row['DR2']
    if haplotype in haplotypes and dr2 >= 0.3:
        for sample in dosages_info.columns[3:]:
            if row[sample] != 0:  # Check if the dosage value is not zero
                for isotype, count in haplotypes[haplotype].items():
                    c4_isotype_counts[sample][isotype] += count

# Convert the dictionary to a DataFrame
c4_isotype_counts_df = pd.DataFrame(c4_isotype_counts).transpose()

# Save the resulting counts to a new file
c4_isotype_counts_df.to_csv('c4_isotype_counts.csv', index=True)
```

```python
for col in dosages_info.columns[3:]:
    dosages_info[col] = pd.to_numeric(dosages_info[col], errors='coerce')

# Initialize a dictionary to store C4 isotype dosages for each sample
c4_isotype_dosages = {sample: {'C4A': 0.0, 'C4B': 0.0, 'C4L': 0.0, 'C4S': 0.0} for sample in dosages_info.columns[3:]}

# Calculate the dosage of each C4 isotype for each sample
for index, row in dosages_info.iterrows():
    haplotype = row['ALT']
    dr2 = row['DR2']
    if haplotype in haplotypes and dr2 >= 0.3:
        for sample in dosages_info.columns[3:]:
            dosage = row[sample]
            if dosage != 0 and not pd.isnull(dosage):  # Check if the dosage value is not zero and is numeric
                #print(f"Processing sample {sample}, haplotype {haplotype}, DR2 {dr2}, dosage {dosage}")
                for isotype, count in haplotypes[haplotype].items():
                    c4_isotype_dosages[sample][isotype] += dosage * count
                    #print(f"  Adding {dosage} * {count} to {isotype} for sample {sample}")

# Convert the dictionary to a DataFrame
c4_isotype_dosages_df = pd.DataFrame(c4_isotype_dosages).transpose()
c4_isotype_dosages_df.to_csv('c4_isotype_dosages_info.csv', index=True)
```
